```python
import random

def play_game():
    # 初始化10个盒子和初始金额
    boxes = [0] * 10
    money = 100

    # 随机放入金额
    winning_box = random.randint(0, 9)
    boxes[winning_box] = money

    # 游戏开始，初始化游戏状态
    current_player = 0
    total_players = 10
    money_collected = 0
    opened_boxes = set()
    open_count = 0  # 初始化盒子打开计数器

    # 游戏主循环
    while money_collected == 0:
        # 当前玩家打开一个盒子
        available_boxes = [box for box in range(10) if box not in opened_boxes]
        if not available_boxes:
            break  # 所有盒子都已打开
        chosen_box = random.choice(available_boxes)

        if boxes[chosen_box] == 0:
            money *= 2  # 如果未找到钱，翻倍金额
        else:
            money_collected = boxes[chosen_box]  # 找到钱，游戏结束
        boxes[chosen_box] = 0  # 盒子已被打开
        opened_boxes.add(chosen_box)
        open_count += 1  # 增加盒子打开计数器

        print(f"第{open_count}次打开：玩家{current_player + 1} 打开盒子{chosen_box + 1}，金额：${money_collected}")

        current_player = (current_player + 1) % total_players

    return money_collected

# 进行1000次循环测试
num_simulations = 1000
total_money_collected = 0

for _ in range(num_simulations):
    money_collected = play_game()
    total_money_collected += money_collected

average_money_collected = total_money_collected / num_simulations

print(f"平均每轮游戏收集的金额：${average_money_collected:.2f}")
```
